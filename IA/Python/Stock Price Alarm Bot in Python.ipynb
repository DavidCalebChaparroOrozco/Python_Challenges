```python
import os
import time
import pandas_datareader as web
from winotify import Notification, audio
```

```python
tickers = ['AAPL', 'FB','NVDA']

upper_limits = [400, 220, 240]
lower_limits = [200, 130, 140]
```

```python
#for ticker in tickers:
#    print(web.DataReader(ticker, 'yahoo').iloc[-1]['Close'])
```

```python
while True:
    last_prices = [web.DataReader(ticker, 'yahoo')['Adj Close'][-1]for ticker in tickers]
    print(last_prices)
    time.sleep(1)
    for i in range(len(tickers)):
        if last_prices[i] > upper_limits[i]:
            toast = Notification(app_id="Caleb Stock Alarm Bot",
                                title='Price Alert For ' + tickers[i],
                                 msg=f"{tickers[i]}  has reached a price of {last_prices[i]:.2f}. You might want to sell.",
                                 icon=os.path.join(os.getcwd(), "buy.jpg"),
                                 duration="long")
            toast.add_actions(label='Go to Stockbroker', launch='https://davidcalebchaparroorozco.netlify.app/#')
            toast.set_audio(audio.LoopingAlarm6, loop=True)
            toast.show()
            
        elif last_prices[i] < lower_limits[i]:
            toast = Notification(app_id="Caleb Stock Alarm Bot",
                                title='Price Alert For ' + tickers[i],
                                 msg=f"{tickers[i]}  has reached a price of {last_prices[i]:.2f}. You might want to buy.",
                                 icon=os.path.join(os.getcwd(), "coin.jpg"),
                                 duration="long")
            toast.add_actions(label='Go to Stockbroker', launch='https://davidcalebchaparroorozco.netlify.app/#')
            toast.set_audio(audio.LoopingAlarm8, loop=True)
            toast.show()
        time.sleep(1)
            
```
